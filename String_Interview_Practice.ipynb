# **Reverse a string each word individually**

```python
text = ("Hello World")
words = text.split()
rev_word = []
for word in words:
  rev_word.append(word[::-1])
result = " ".join(rev_word)
print(result)
```

```python
s = "Rohan Rathore"
rev_st = " ".join(word[::-1] for word in s.split())
print(rev_st)
```

```python
s = "Rohan Rathore"
rev_st = " ".join(map(lambda x:x[::-1],s.split()))
print(rev_st)
```

# **Write a Python program to check whether a given string is a palindrome or not.**

```python
def is_palindrom(name):
  return name == name[::-1]

name = "madam"
if is_palindrom(name):
  print("Palindrom")
else:
  print("Not Palindrom")
```

# **Count the Occurrence of Each Character in a String**

```python
story = ("Programming")
for i in story:
  print(f"{i}",story.count(i))
```

```python
from collections import Counter

text = "programming"
count = Counter(text)

for k, v in count.items():
    print(k, ":", v)
```

# **Check if Two Strings are Anagrams**

```python
def anagram(s1,s2):
  return sorted(s1) == sorted(s2)

s1 = ("listen")
s2 = ("silent")

if anagram(s1,s2):
  print("Anagram")
else:
  print("Not Anagram")
```

# **Without using any in-built function:**

```python
text = ("hello world")
A = "abcdefghijklmnopqrstuvwxyz"
B = "ABCDEFGHIJKLMNOPQRSTUVWXYZ"
new = ""
for j in text:
  found = False
  for i in range(26):
    if j == A[i]:
      new += B[i]
      found = True
      break
  if not found:
    new += j
print(new)
```
